## 2. Phân cụm (K-Means Clustering)

```
cluster_model = MachineClustering(random_state=42)
X_scaled_cluster = cluster_model.prepare_data(df_cleaned)
```

### Tìm số cụm K tối ưu

```
best_k = cluster_model.find_optimal_k(X_scaled_cluster, max_k=8, plot_elbow=True)
```

### Gán nhãn cụm và Profiling rủi ro

```
df_clustered = cluster_model.fit_predict(df_cleaned, n_clusters=4) # Sử dụng K=4 mặc định
profile = cluster_model.cluster_profiling(df_clustered)
profile
```
