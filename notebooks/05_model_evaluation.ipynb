## Mở rộng: Phân tích kết quả và Insights

1. **Quy trình bảo trì**: Dựa vào luật kết hợp (Association Rules), có thể thấy các máy có dấu hiệu Nhiệt độ quá trình (Process temperature) cao và mô-men xoắn (Torque) lớn dễ bị hỏng vì lỗi OSF (Overstrain Failure).
2. **Phân nhóm K-Means**: Phân cụm giúp phân loại máy thành các nhóm có mức độ lỗi khác nhau. Cụm có tỷ lệ lỗi cao (>10%) nên đưa vào theo dõi gắt gao. Có thể áp dụng để đánh giá tình trạng thiết bị sau mỗi ca làm việc.
3. **Sức mạnh mô hình phân lớp**: XGBoost hoạt động rất tốt đối với bài toán mất cân bằng kết hợp với SMOTE. Model sử dụng đặc trưng `Torque`, `Tool wear` làm feature quan trọng lớn nhất thay vì Temperature.
4. **Giá trị của nhãn giả (Semi-supervised)**: Kỹ thuật Self-Training giúp tăng độ ổn định của model nếu thiết lập ngưỡng `threshold` đủ lớn (ví dụ 0.85). Nếu `threshold` qua thấp sẽ sinh ra rất nhiều False Positive Pseudo-label.
5. **Hồi quy Tool Wear**: Vì đặc trưng Tool Wear tăng dần theo chuỗi vận hành của đơn vị thời gian (UDI), sử dụng phương pháp tạo đặc trưng trễ (Lag >= 1) đối với `Tool wear` đem lại MAPE và RMSE rất nhỏ.
